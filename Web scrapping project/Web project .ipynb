```python
import requests
from bs4 import BeautifulSoup

# Fetch the HTML content from the URL
url = "https://en.wikipedia.org/wiki/List_of_Nobel_laureates"
res = requests.get(url)

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(res.text, 'html.parser')

# Find the first table in the HTML content
table = soup.find_all('table')[0]

# Extract the table rows
table_rows = table.find_all('tr')

# Write the data to CSV
with open('nobel_data.csv', mode='w', encoding='utf-8') as fd:
    # Write the header
    fd.write('Year, Physics, Chemistry, Physiology or Medicine, Literature, Peace, Economics (The Sveriges Riksbank Prize)\n')
    print (f"data {len(table_rows)} sacrapped ")
    # Iterate over the rows
    row_counter = 0 
    for row in table_rows[1:]:
        row_counter += 1
        cells = row.find_all('td')
        row_data = []
        
        #if row_counter in range(40,43) and cells != numeric :
            #row_data.append("Cancelled nobel for world war II")
    
       
        # Extract cell texts
        for cell in cells:
               
            row_data.append(cell.text.strip().replace('\n', ', '))
                    
            if row_counter in range(40,43) :
                row_data.extend(['Cancelled due to World War II'] * 6)
           
                       
        
           
        print(f'Row {row_counter} scrapped ')
            
        print(f' Cell {len(row_data)} extracted \n')
            
        # Handle rows len <  7 and joining 
        while len(row_data) < 7:
            row_data.append('')
        
       
        fd.write(', '.join(row_data) + '\n')
        
## View CSV file using PANDAS 

print("Data saved to nobels.csv")
```
